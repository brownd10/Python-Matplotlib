```python
# Create dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load in csv
cities = pd.read_csv('Resources/city_data.csv')
riders = pd.read_csv('Resources/ride_data.csv')
```

```python
# Display dataframe shape for city_info
cities.shape
```

```python
# Display dataframe shape for ride_info
riders.shape
```

```python
# Merge dataframes on 'city' so that no data is lost
combined_data = pd.merge(riders, cities, how='outer', on='city')
combined_data
```

```python
# Check shape of combined dataframes
combined_data.shape
```

```python
# Check for any missing values in each dataset
combined_data.isnull().sum()
```

```python
# Calculate variables

# Average fare ($) per city
combined_data.fare.mean()
```

```python
# Display average fare per city
avg_city_fare = combined_data.groupby('city').fare.mean()
avg_city_fare
```

```python
# Total number of rides per city
combined_data.ride_id.count()
```

```python
# Display total number of rides per city
total_city_rides = combined_data.groupby('city').ride_id.count()
total_city_rides
```

```python
# Total number of drivers per city
combined_data.driver_count.sum()
```

```python
# Display total number of drivers per city
driver_total = combined_data.groupby('city').driver_count.mean()
driver_total
```

```python
# Create bubble plot
plt.scatter(total_city_rides, avg_city_fare, marker="o", facecolors="blue", edgecolors="grey", s=total_city_rides*20, alpha=0.50)

plt.ylim(0, 50)
plt.xlim(0, 40)


#plt.legend(('Urban', 'Suburban', 'Rural'), scatterpoints=1, loc='upper right', ncol=2, fontsize=6)

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

plt.show()
```
